### Outlier Removal
###### Since the data distribution here is mostly skewed, and not normally distributed, hence Inter-Quartile Range (IQR) proximity rule is applicable

```python
f,ax=plt.subplots(figsize=(10,5))
sns.boxplot(df_clean1['balance'])
```

```python
# Outlier detection using percentile technique 5% to 95%

balance_low_lim=df_clean1['balance'].quantile(0.05)
balance_upp_lim=df_clean1['balance'].quantile(0.95)

df_clean2=df_clean1.copy()


df_clean2['balance'].replace(to_replace=df_clean2['balance'][df_clean2['balance']<balance_low_lim].tolist(), value=balance_low_lim, inplace=True)
df_clean2['balance'].replace(to_replace=df_clean2['balance'][df_clean2['balance']>balance_upp_lim].tolist(), value=balance_upp_lim, inplace=True)

print(balance_upp_lim)
print(balance_low_lim)
```

###### Removing the Outliers of 'duration'

```python
# Outlier detection using 3 sigma technique

df_clean4=df_clean2.copy()

dur_upp=df_clean4['duration'].mean()+3*df_clean4['duration'].std()
dur_low=df_clean4['duration'].mean()-3*df_clean4['duration'].std()



df_clean4['duration'].replace(to_replace=df_clean4['duration'][df_clean4['duration']>=dur_upp].tolist(), value=df_clean4['duration'].mean(), inplace=True)
df_clean4['duration'].replace(to_replace=df_clean4['duration'][df_clean4['duration']<=dur_low].tolist(), value=df_clean4['duration'].mean(), inplace=True)
```

```python
f,ax=plt.subplots(figsize=(10,5))
sns.boxplot(df_clean4['duration'])
```

##### Removing the Outliers of 'campaign'

```python
df_clean5=df_clean4.copy()

f,ax=plt.subplots(figsize=(10,5))
sns.boxplot(df_clean5['campaign'])
```

```python
df_clean5['campaign'].describe()
```

```python
# Outlier Detection using IQR Technique


campaign_per75=df_clean5['campaign'].quantile(0.75)
campaign_per25=df_clean5['campaign'].quantile(0.25)

campaign_IQR = campaign_per75 - campaign_per25

campaign_upp_lim = campaign_per75 + 1.5*campaign_IQR
campaign_low_lim = campaign_per25 - 1.5*campaign_IQR

df_clean5['campaign'].replace(to_replace=df_clean5['campaign'][df_clean5['campaign']>=campaign_upp_lim].tolist(), value=df_clean5['campaign'].mean(), inplace=True)
df_clean5['campaign'].replace(to_replace=df_clean5['campaign'][df_clean5['campaign']<=campaign_low_lim].tolist(), value=df_clean5['campaign'].mean(), inplace=True)
```

```python
f,ax=plt.subplots(figsize=(10,5))
sns.boxplot(df_clean5['campaign'])
```

```python
import math
def ILATE(col):
    fn=['inv', 'log', 'sqr', 'cub', 'sqrt', 'cbrt', 'sin', 'cos', 'exp']
    for i in fn:
        x=col+'_'+i
        if i == 'inv':
            df_clean5[x]=1/df_clean5[col]
        elif i == 'log':
            df_clean5[x]=np.log(df_clean5[col])
        elif i == 'sqr':
            df_clean5[x]=df_clean5[col]**2
        elif i == 'cub':
            df_clean5[x]=df_clean5[col]**3
        elif i == 'sqrt':
            df_clean5[x]=df_clean5[col]**(0.5)
        elif i == 'cbrt':
            df_clean5[x]=df_clean5[col]**(1/3)
        elif i == 'sin':
            df_clean5[x]=np.sin(df_clean5[col])
        elif i == 'cos':
            df_clean5[x]=np.cos(df_clean5[col])
        else:
            df_clean5[x]=df_clean5[col]**(1/1.2)
            
    return df_clean5
ILATE('balance')
ILATE('duration')
df_clean5.columns
```
